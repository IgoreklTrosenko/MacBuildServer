```python
def filtr(x):
    a = ''
    for i in str(x):
        if i in ('1234567890'):
            a=a+str(i)
    return str(a)
print(filtr('818a8ae3-4d58-11ed-907e-848f69e142cf'))
```

```python
dfleads['lead_created_at']=pd.to_datetime(dfleads['lead_created_at'])
dfleads['lead_id']=dfleads['lead_id'].apply(filtr)
dfleads['client_id']=dfleads['client_id'].apply(filtr)
dfleads=dfleads.drop_duplicates()
dfleads.head(1)
```

```python
# new_df.drop(columns=[['d_utm_source','d_utm_source','d_utm_medium','created_at','d_lead_utm_source','d_lead_utm_medium']],axis=1)
dff =new_df[['lead_created_at','d_utm_campaign','d_lead_utm_campaign','d_utm_content','d_lead_utm_content','m_clicks','m_cost','lead_id','d_lead_utm_term','client_id','leads']]
dff.head(1)
```

```python
purchases.head(1)
```

```python
daraframe=pd.merge(dff,purchases, how ='outer', on = 'client_id')
daraframe['d_utm_campaign']=daraframe['d_utm_campaign']+daraframe['d_lead_utm_campaign']
daraframe=daraframe.drop(columns=['d_lead_utm_campaign'])
daraframe['d_utm_content']=daraframe['d_utm_content']+daraframe['d_lead_utm_content']
daraframe=daraframe.drop(columns=['d_lead_utm_content'])
daraframe=daraframe.drop(columns=['d_lead_utm_term'])



daraframe.info()
```

```python
daraframe.head(1)
```

```python
from datetime import datetime, timedelta
delta15 =timedelta(days=15)
delta0 =timedelta(days=0)
daraframe['delta_time']=daraframe['purchase_created_at']-daraframe['lead_created_at']
daraframe=daraframe[(daraframe['delta_time']<=delta15) &(daraframe['delta_time']>=delta0)]
daraframe['medium']='cpc'
daraframe['source'] ='yandex'
daraframe['m_clicks']=daraframe['m_clicks'].fillna(0).astype(int)
```

```python
daraframe.info()
```

Здесь мы попробуем выделить лиды, которые мы засчитаем, тоесть которые выполняют условия, последнему лиду зассчитывается близжайшие продажи и оставим в таблице только те данные, которые удовлетворяют нашим требованиям

```python
dataframe= daraframe.groupby(['purchase_created_at','lead_id','purchase_id','client_id','d_utm_campaign','medium','source','m_clicks','m_cost','m_purchase_amount','leads'])['delta_time'].min().reset_index()
dataframe
```

# Количество кликов	Расходы на рекламу	Выручка от продаж

```python
dataframe.groupby(['purchase_created_at','d_utm_campaign','lead_id','client_id','purchase_id','medium','source'])[['m_clicks','m_cost','m_purchase_amount','leads']].sum().reset_index()
dataframe.rename(columns={'m_clicks': 'Количество кликов','m_cost': 'Расходы на рекламу','leads': 'Количество лидов','m_purchase_amount':'Выручка от продаж'}, inplace =True)
```

```python
dataframe['ROAS']=dataframe['Выручка от продаж']/dataframe['Расходы на рекламу']
```

```python
dataframe=dataframe[['purchase_created_at','d_utm_campaign','d_utm_campaign','source','Количество кликов','Расходы на рекламу','Количество лидов','Выручка от продаж','ROAS']].copy()
dataframe
```

```python
dataframe.to_csv("dataframe1.csv", index=False)
```
